```python
import pandas as pd
unemployement_data = pd.read_csv("Unemployment in India.csv")
```

```python
print(unemployement_data)
print("===== Missing Value =====")
print(unemployement_data.isnull().sum())
print("===== Summary statistics =====")
print(unemployement_data.describe())
```

```python
#cleaning data by droping the rows that is missing
unemployed_new = unemployement_data.dropna()
print(unemployed_new.shape) 
```

```python
print (unemployed_new.head())
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(12, 6))
plt.plot(unemployed_new.index, unemployed_new["Estimated Unemployment Rate (%)"], marker="o", linestyle="-", color="b")
plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.grid()
plt.show()
```

```python
print(unemployed_new.columns)
```

```python
unemployed_new.columns = unemployed_new.columns.str.strip()
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x="Region", y="Estimated Unemployment Rate (%)", data=unemployed_new, hue="Region", palette="coolwarm", legend=False)
plt.xticks(rotation=90)
plt.title("Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate (%)")
plt.show()

```
